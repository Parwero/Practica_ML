```python
# Lista de variables de reseñas secundarias (las de escala 0-10)
# Esto lo tuve que corregir en vializacion porque me sumo las de 100 y 10 lo que me daba 160 de maxima puntuacion 
cols_reviews_secundarias = [
    'Review Scores Accuracy',
    'Review Scores Cleanliness',
    'Review Scores Checkin',
    'Review Scores Communication',
    'Review Scores Location',
    'Review Scores Value'
]

# 1. Imputamos los nulos en la variable PRINCIPAL 'Review Scores Rating' (escala 0-100)
# Usamos la media de train para rellenar
rating_mean = train['Review Scores Rating'].mean()
train['Review Scores Rating'] = train['Review Scores Rating'].fillna(rating_mean)
test['Review Scores Rating'] = test['Review Scores Rating'].fillna(rating_mean)

# 2. Eliminamos las variables secundarias (altamente correlacionadas y con otra escala)
# Así evitamos el ruido y la redundancia.
train = train.drop(cols_reviews_secundarias, axis=1)
test = test.drop(cols_reviews_secundarias, axis=1)

print("Variables de reseñas secundarias eliminadas.")
print("Mantenemos 'Review Scores Rating' como indicador principal.")

# Verificamos
print(train['Review Scores Rating'].describe())
```

```python
# Sigo imputando columnas, ahora ire a nathrooms, bedrooms, beds, y host, aunque tienen pocos valores nulos, al ser numericos puedo imputarlso con la media
```

```python

# Lista de variables específicas que suelen tener huecos
cols_estructurales = ['Bathrooms', 'Bedrooms', 'Beds']
cols_host = ['Host Response Rate', 'Host Listings Count', 'Calculated host listings count']

# Imputamos con la mediana
for col in cols_estructurales:
    if col in train.columns:
        mediana = train[col].median()
        train[col] = train[col].fillna(mediana)
        test[col] = test[col].fillna(mediana) 
        print(f"  - {col}: Imputado con {mediana}")

# Imputamos Host
for col in cols_host:
    if col in train.columns:
        media = train[col].mean()
        train[col] = train[col].fillna(media)
        test[col] = test[col].fillna(media)
        print(f"  - {col}: Imputado con {media:.2f}")
```

```python
# Con calendar queria hacer targetr enconder pero no le vi mucho sentido, creo que lo mas optimo era cambiar el texto por numeros, tipo hoy es 1, 1 week 7, 
def parse_calendar_updated(val):
    val = str(val).lower()
    
    if 'today' in val:
        return 1
    
    elif 'yesterday' in val:
        return 1
    
    
    elif 'never' in val:
        
        return 3650 
    # este valor lo puse para que como tope de fecha, fue mas complicado de lo que pensaba y tuve que corregir un monton de cosas 
    
    else:
        try:
            parts = val.split()
            if parts[0] == 'a':
                number = 1
            else:
                number = int(parts[0])
            
            if 'week' in val:      
                return number * 7
            elif 'month' in val:  
                return number * 30 
            elif 'day' in val:     
                return number
            else:
               
                return np.nan
        except:
           
            return np.nan

if 'Calendar Updated' in data_airbnb.columns:
    data_airbnb['Calendar Updated'] = data_airbnb['Calendar Updated'].apply(parse_calendar_updated)

mean_days = data_airbnb['Calendar Updated'].mean()
data_airbnb['Calendar Updated'] = data_airbnb['Calendar Updated'].fillna(mean_days)

print(data_airbnb['Calendar Updated'].head(10))
print(data_airbnb['Calendar Updated'].dtype)
```

```python
import pandas as pd
import numpy as np
import warnings
from sklearn.model_selection import train_test_split

warnings.filterwarnings('ignore')

# Carga de contenido
df = pd.read_csv('./data/airbnb-listings-extract.csv', sep=';', decimal='.')

# Aqui quite dolar y , porque es
df['Price'] = df['Price'].astype(str).str.replace('$', '').str.replace(',', '')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df = df.dropna(subset=['Price'])

#  Elimino columnas
cols_to_drop = [
    'ID', 'Scrape ID', 'Host ID', 'Listing Url', 'Last Scraped', 'Name', 
    'Summary', 'Description', 'Host Name', 'Picture Url', 'Host URL', 
    'Host Thumbnail Url', 'Host Picture Url', 
    'Thumbnail Url', 'Medium Url', 'XL Picture Url', 
    'Street', 'Host About', 'Host Location', 'Host Neighbourhood',
    'Geolocation', 'Space', 'Amenities', 'House Rules', 'Notes', 
    'Transit', 'Access', 'Interaction', 'Neighborhood Overview',
    'Experiences Offered'
]

# Borramos las que existan
df = df.drop([c for c in cols_to_drop if c in df.columns], axis=1)

# División Train/Test 
train, test = train_test_split(df, test_size=0.2, shuffle=True, random_state=0)

print(f"Train shape: {train.shape}")
```

```python
# Ahora trabajasmos con el dataset

# Hago la limpieza de otras columnas de dinero 
money_cols = ['Security Deposit', 'Cleaning Fee', 'Extra People']
for col in money_cols:
    # Train
    train[col] = train[col].astype(str).str.replace('$', '').str.replace(',', '')
    train[col] = pd.to_numeric(train[col], errors='coerce').fillna(0)
    # Test
    test[col] = test[col].astype(str).str.replace('$', '').str.replace(',', '')
    test[col] = pd.to_numeric(test[col], errors='coerce').fillna(0)

# Elimino de fechas como en clase
date_cols = ['Host Since', 'First Review', 'Last Review', 'Calendar last Scraped', 'Calendar Updated']
train = train.drop([c for c in date_cols if c in train.columns], axis=1)
test = test.drop([c for c in date_cols if c in test.columns], axis=1)

# Imputo los nulos
num_cols = train.select_dtypes(include=[np.number]).columns
for col in num_cols:
    if col != 'Price':
        median_val = train[col].median()
        train[col] = train[col].fillna(median_val)
        test[col] = test[col].fillna(median_val)

# Codifico
cat_cols = train.select_dtypes(include=['object']).columns
target = 'Price'

for col in cat_cols:
# Calculo la media del precio en train
    mean_price = train.groupby(col)[target].mean()
# Aplicao a train
    train[col] = train[col].map(mean_price)
    train[col] = train[col].fillna(train[target].mean())    
# Aplico test
    test[col] = test[col].map(mean_price)
    test[col] = test[col].fillna(train[target].mean())

# Genero las caracteristicas
train['Beds_Bath'] = train['Beds'] * train['Bathrooms']
test['Beds_Bath'] = test['Beds'] * test['Bathrooms']

print(train.head(3))
```
